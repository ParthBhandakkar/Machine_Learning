```python
train_df_significant = train_df.drop(['id', 'product_code', 'attribute_0', 'attribute_1'], axis = 1)
X_train=train_df_significant.drop('failure',axis=1)       # Independent variables
Y_train=train_df_significant['failure']
```

```python
from sklearn.preprocessing import StandardScaler

X_train_S = StandardScaler().fit_transform(X_train)
```

```python
print(X_train_S.shape)
print(Y_train.shape)
```

```python
test_df = pd.read_csv('test_playground.csv')
test_df
```

```python
test_df_significant = test_df.drop(['id', 'product_code', 'attribute_0', 'attribute_1'], axis = 1)
X_test=test_df_significant
X_test
```

```python
X_test.isnull().values.any()
```

```python
for i in X_test.columns[X_test.isnull().any(axis=0)]:    
    X_test[i].fillna(X_test[i].mean(),inplace=True)
X_test.isnull().values.any()
```

```python
X_test_S = StandardScaler().fit_transform(X_test)
X_test_S.shape
```

```python
from keras.models import Sequential
from keras.layers import Dense, Activation, LeakyReLU
from keras import optimizers
```

```python
NN_model_Classifier = Sequential()

# The Input Layer :
NN_model_Classifier.add(Dense(128, kernel_initializer='normal',input_dim = X_train.shape[1], activation='relu'))

# The Hidden Layers :
NN_model_Classifier.add(Dense(64, kernel_initializer='normal',activation='relu'))  # sigmoid, tanh

NN_model_Classifier.add(Dense(32, kernel_initializer='normal'))
NN_model_Classifier.add(LeakyReLU(alpha=0.1))

NN_model_Classifier.add(Dense(16, kernel_initializer='normal'))
NN_model_Classifier.add(LeakyReLU(alpha=0.1))


# The Output Layer :
NN_model_Classifier.add(Dense(9, kernel_initializer='normal',activation='sigmoid'))  # except softmax

# Compile the network :
NN_model_Classifier.compile(loss='mean_absolute_error', optimizer='adam', metrics=['accuracy'])
NN_model_Classifier.summary()
```

```python
EPOCH=100
Network_Classifier=NN_model_Classifier.fit(X_train_S, Y_train, epochs=EPOCH, batch_size=200)
```

```python
y_predicted = NN_model_Classifier.predict(X_test_S)
# y_predicted_list = y_predicted.tolist()
# y_pred = y_predicted_list[0]
# y_pred_final = []
# for i in y_pred:
#     j = i*10000000000000
#     y_pred_final.append(j)
# len(y_pred)
len(y_predicted)
```
